# 1. 집값 데이터 읽기

```python
import numpy as np

#(집 크기(평)), 가격(억원)
# 학교 근처(홍은홍제구기신영평창녹번동)

import pandas as pd
data=pd.read_csv('house_price.txt', names=['size','price']) ##txt or csv 파일, csv는 ,로 나열
##names: index를 붙여준다.
print data
X=data['size']  #제곱미터(평)
y=data['price'] # 억원
m=len(data)     #정보 개수(행 개수) m=52
```

```python
#numpy array형태로 변환, 형태 변환(m)->(m,1)
X=(np.array(X)).reshape(m,1)
y=(np.array(y)).reshape(m,1)
print X.shape, y.shape
```

# 2. 그래프 그리기

```python
import matplotlib.pyplot as plt ##plt 함수 사용
plt.plot(X, y, 'b.') ##X(평수), y(가격), b.:블루의 .으로 표현해줘
plt.xlabel("Size of a house in Pyeong)") ##집 크기(평) (한글로 쓰면 에러뜸)
plt.ylabel("Price in uck-won")           ##매매가(억원)
plt.show()
```

# 3. Gradient descent -> theta

```python
X_b=np.c_[np.ones((m,1)),X] ##모든 샘플에 x0=1을 추가-->np.ones((m,1))과 X합침
#c_ : concentration. 배열을 옆으로 붙이기

learning_rate=0.0001 #학습률
n_iter=200

theta=np.zeros((2,1))
gradients=np.zeros((2,1))

for i in range(n_iter):
    #코드작성
    gradients=(1.0/m)*np.dot(X_b.T,(np.dot(X_b,theta)-y)) ##1아니고 1.0으로 해야함!!
    theta=theta-learning_rate*gradients
print "theta:"
print theta
```

# 4. 집 크기 최소(10평)~최대(90평)*theta -> 그래프 그리기

```python
X_new=np.array([[10],[90]]) #10~90평
X_new_b=np.c_[np.ones((2,1)),X_new] #모든 샘플에 x0=1을 추가
y_predict=X_new_b.dot(theta)
print "10평과 90평 집의 예측 가격"
print y_predict
```

```python
plt.plot(X_new,y_predict,'r-')
plt.plot(X,y,'b.')
plt.xlabel("Size of a house in Pyeong)") ##집 크기(평)
plt.ylabel("Price in uck-won")           ##매매가(억원)
plt.show()
```

# 5. 임의의 집 크기(x_my)에 대한 집값 예측하기

```python
X_mine=np.array([[25]]) #우리집이 25평이라면 얼마?
X_mine_b=np.c_[np.ones((1,1)),X_mine] #모든 샘플에 x0=1을 추가
y_predict=X_mine_b.dot(theta)
print y_predict, "억원"
```

# 6. 확률적 경사 하강법 함수 호출하여 theta 구하기 (선택) 


```python
from sklearn.linear_model import SGDRegressor
sgd_reg=SGDRegressor(n_iter=100, penalty=None, eta0=0.0001) #eta0(learning rate), 하이퍼파라미터
sgd_reg.fit(X,y.ravel()) #sgd_reg.fit(X,y.ravel()) #1차원 배열로 만들기 (m,1)->(m)
```

```python
print "theta:"
print sgd_reg.intercept_, sgd_reg.coef_ # theta0,1

y_pred=sgd_reg.predict(X) ##sgd는 랜덤하게 데이터 한개를 고른다.
plt.plot(X,y,'b.')
plt.plot(X,y_pred,'r')
plt.show()
```

```python
#우리집이 25평이라면 얼마?
y_pred=sgd_reg.predict([[25]])
print y_pred, "억원"
```

# 7. 비용함수 J(세타) 3차원 그래프로 그리기 (선택)

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ComputeJ(t0,t1,X,y):
    m=len(X)
    X_b=np.c_[np.ones((m,1)),X] ##모든 셈플에 x0=1 추가
    theta=np.array([t0,t1])
    cost_vector=X_b.dot(theta)-y
    cost_vector=cost_vector.reshape(-1)
    
    #각 원소 제곱
    cost2=np.square(cost_vector)
    
    #원소를 더하기
    cost_sum=np.sum(cost2)/(2*m)
    
    return cost_sum
```

## 1. theta0, theta1, J값 구하기

```python
d=100 #of split
s=1.0

theta0=np.linspace(-15*s,15*s,d)
theta1=np.linspace(-1.0*s, 1.0*s,d)

J=np.ones((d,d))
for i, t0 in enumerate(theta0):
    for j, t1 in enumerate(theta1):
        cost=ComputeJ(t0, t1, X, y)
        J[i,j]=cost
```

## 2. 그리기

```python
pX, pY=np.meshgrid(theta0, theta1)

fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(pX,pY,J,300,cmap='viridis')
ax.set_xlabel('theta0')
ax.set_ylabel('theta1')
ax.set_zlabel('J')
plt.show
```
